**Problems in Batch Gradient Descent:**
1. Memory inefficieny
2. Better convergence

**Solution:** Rather than loading the entire data, we should load using batches to train the Model.

**Two Classes provided by Pytorch: Dataset and DataLoader.**
->Using these 2 classes,we can apply minibatch Gradient Descent

```python
from sklearn.datasets import make_classification
import torch
```

```python
X, y = make_classification(
    n_samples=10,
    n_features=2,
    n_informative=2,
    n_redundant=0,
    n_classes=2,
    random_state=42

)
```

```python
X = torch.tensor(X, dtype=torch.float32)
y = torch.tensor(y, dtype=torch.long)
```

```python
X
```

```python
y
```

```python
from torch.utils.data import Dataset, DataLoader
```

```python
class CustomDataset(Dataset):

  def __init__(self, features, labels):

    self.features = features
    self.labels =labels

  def __len__(self):

    return self.features.shape[0]

  def __getitem__(self, index):

    return self.features[index], self.labels[index]
```

```python
dataset = CustomDataset(X, y)
```

```python
len(dataset)
```

```python
dataset[7]
```

```python
dataloader = DataLoader(dataset, batch_size=2, shuffle=False)

#This shuffling is done by the sampler present inside dataloader
```

```python
for batch_features, batch_labels in dataloader:

  print(batch_features)
  print(batch_labels)
  print("-"*50)
```
